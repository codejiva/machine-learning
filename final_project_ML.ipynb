```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
imputed_data = imputer.fit_transform(dataset)
dataset = pd.DataFrame(imputed_data, columns=dataset.columns)

```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(dataset[['X1']])
normalized_df = pd.DataFrame(scaled_data, columns=['X1'])
```

```python
plt.hist(normalized_df)
```

```python
dataset['X1'] = np.log1p(dataset['X1'])
```

```python
plt.hist(dataset['X1'])
```

Normalisasi tidak mengubah distribusi. Normalisasi hanya mengubah skala nilai dalam suatu fitur, tetapi tidak mengubah secara mendasar distribusi data yang ada. Jika data memiliki kemiringan positif (lebih banyak titik data di ujung bawah), normalisasi akan tetap mempertahankan bentuk yang condong tersebut, hanya dalam rentang baru (misalnya, 0-1).

```python
# kita asumsikan X1 dan X2 terdapat interaksi

dataset['X1_X2_interaction'] = dataset['X1'] * dataset['X2']
dataset['X1_to_X2_ratio'] = dataset['X1'] / dataset['X2']

print(dataset['X1_X2_interaction'])
print(dataset['X1_to_X2_ratio'])
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X = dataset.drop('X1', axis=1)
y = dataset['X1']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# train model
model = LinearRegression()
model.fit(X_train, y_train)

# bikin model prediksi
y_pred = model.predict(X_test)

# Evaluation Metrics
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt


features_to_use = [col for col in dataset.columns if col != 'X1']
X = dataset[features_to_use]
y = dataset['X1']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train a Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# bikin model prediksi
y_pred = model.predict(X_test)

# Evaluation Metrics
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)  # ngitung RMSE
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("Mean Absolute Error:", mae)
print("R-squared:", r2)

# Visualizations
plt.scatter(y_test, y_pred)
plt.xlabel("Actual X1 Value")
plt.ylabel("Predicted X1 Value")
plt.title("Predicted vs. Actual X1 Value")
plt.show()

# Residual Plot
plt.scatter(y_pred, y_test - y_pred)
plt.xlabel("Predicted X1 Value")
plt.ylabel("Residuals")
plt.title("Residuals vs. Predicted X1 Value")
plt.show()
```

Berdasarkan analisis regresi yang dilakukan, berikut adalah kesimpulan yang dapat ditarik:

1. Prediksi Akurat: Model regresi yang dibangun menunjukkan kemampuan prediksi yang sangat akurat untuk nilai "X1" berdasarkan fitur-fitur lainnya dalam data. Hal ini dibuktikan dengan nilai-nilai metrik evaluasi yang sangat baik:

Mean Squared Error (MSE): 7.54e-06 (sangat kecil)
Root Mean Squared Error (RMSE): 0.0027 (sangat kecil)
Mean Absolute Error (MAE): 0.0021 (sangat kecil)
R-squared: 0.964 (sangat tinggi)
2. Hubungan Kuat: Hasil analisis menunjukkan hubungan yang kuat antara fitur-fitur yang digunakan dan nilai "X1". Model regresi berhasil menangkap hubungan ini dan menghasilkan prediksi yang akurat.

3. Keandalan Model: Metrik-metrik yang diperoleh menunjukkan bahwa model regresi yang dibangun cukup andal dalam memprediksi nilai "X1".


Analisis regresi menunjukkan bahwa model yang dibangun mampu memprediksi nilai "X1" dengan akurasi tinggi. Hal ini menunjukkan bahwa model tersebut dapat menjadi alat yang berharga untuk berbagai aplikasi yang melibatkan prediksi "X1" berdasarkan fitur-fitur lainnya.
